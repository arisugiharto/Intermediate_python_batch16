# Pie Chart

```python

from matplotlib import pyplot as plt
```

```python
slices = [60, 40]
```

```python
plt.pie(slices)

plt.title("My Pie Chart")
# plt.tight_layout()
plt.show()
```

```python
slices = [60, 40]
labels = ['60%', 'Forty']

plt.pie(slices, labels=labels)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

The value shouldn't 100 in total

```python
slices = [60, 40]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [60, 40, 30, 10] # >100 = 60+40+30+10 = 140
#(60/140) x 100% = 42 %
# (40/140) x 100% = 
labels = ['Sixty', 'Forty', 'Thirty', 'Ten']
#colors = ['blue', 'red', 'yellow', 'green']
colors = ['r', 'y', 'b', 'g']

plt.pie(slices, labels=labels, colors=colors,
        wedgeprops={'edgecolor': 'black'})

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

It is not good to use pie chart for data that has more than 5 items

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices, labels=labels)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

### Ephasize an items

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.15, 0] # Menampilkan pie chart yg terbelah

plt.pie(slices, labels=labels, explode=explode)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

### Add shadow

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True) # membuat bayangan dari pie chart

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

### Start at certain point

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, startangle=30) # di mulainya belahan pie chart 

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

### add percentage

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, startangle=90, autopct = '%1.2f%%')#autopct = auto percentage, 
#1.2%%= 2 digit angka di belakang koma

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```
