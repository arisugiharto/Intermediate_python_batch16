# Time Series


```python
from datetime import datetime
import matplotlib.dates as mpl_dates
from matplotlib import pyplot as plt
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y)

plt.tight_layout()
plt.show()
```

```python
plt.plot_date(dates, y, linestyle = 'solid', marker = None)

plt.tight_layout()
plt.show()
```

```python
plt.plot_date(dates, y, linestyle = 'solid')
plt.gcf().autofmt_xdate() # meruah jarak tanggal agar tidak terlalu berdekatan

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=[8,5])

plt.plot_date(dates, y, linestyle = 'solid')
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%B, %d %Y')

plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```
