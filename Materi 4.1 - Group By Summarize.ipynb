# Group By Summarize

For each start location, find the mean distance of the trip

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("My Uber Drives - 2016.csv")
data
```

```python
data["DISTANCE"] = np.where(data["MILES*"] >5, 'Longtrip', 'Short Trip')
data
```

```python
data["YEAR"]=np.array("2021")
data
```

```python
data["YEAR"] = ("2020")
data
```

```python
data["START*"].value_counts()
```

```python
data["START*"].value_counts().head(10)
```

```python
data["START*"].value_counts().tail(10)
```

```python
data["TIME_CAT"] = np.where(data["START_DATE*"]>"1/06/2016 14:00:00", "New_Trip", "Old_Trip")
data
```

```python
#Untuk menampilkan aggregasi mean (rata)dari suatu data
data.groupby("START*")["MILES*"].agg(["mean"])
```

```python
#Jika hanya ingin menampilkan beberapa data saja :
data.groupby("START*")["MILES*"].agg(["mean"]).head(10)
```

```python
#Jika ingin menampilkan >2 aggregasi dalam 1 syntax:
data.groupby("START*")["MILES*"].agg(["mean","sum"]).head(10)
```
