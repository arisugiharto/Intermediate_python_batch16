5. Selesaikan masalah missing value dengan cara berikut:
    - Drop kolom-kolom yang mengandung missing value lebih dari 50% keseluruhan data.
    - Isi missing value pada kolom country dan children dengan data yang paling banyak muncul.
    - isi missing value pada kolom agent dengan mean.
### Jawaban No. 5

```python
df5=df
df5
```

```python
df9 = df5.replace({"?":np.nan, "--":np.nan})
df9
```

```python
df9.isnull().sum()
```

Drop kolom-kolom yang mengandung missing value lebih dari 50% keseluruhan data.

```python
df9.drop("company", axis=1, inplace=True)
```

```python
df9.isnull().sum()
```

Isi missing value pada kolom  children[10] dan country[13] dengan data yang paling banyak muncul.

```python
df10=df9
```

```python
df10.iloc[:, [10,13]].mode()
```

```python
df10.iloc[:, 10]=df10.iloc[:, 10].fillna(df10.iloc[:, 10].mode()[0])
```

```python
df10.iloc[:,13]=df10.iloc[:,13].fillna(df10.iloc[:,13].mode()[0])
```

```python
df10.isnull().sum()
```

Mengganti kolom agent [23] dengan mean

```python
df10.iloc[:,23]=df10.iloc[:,23].fillna(df10.iloc[:,23].mean())
```

```python
df10.isnull().sum()
```

8. Visualisasikan arrival_date_month untuk melihat bulan dimana pengunjung terbanyak
### Jawaban No. 8


```python
df8=df
df8
```

```python
df8.info()
```

```python
df8.iloc[:, [4,9,10,11]]
```

```python
arrival_date_month = df8["arrival_date_month"]
adults = df8["adults"]

plt.hist(arrival_date_month, bins =10, edgecolor="k", orientation = "horizontal")
plt.title("Total visitor in month")
plt.ylabel("Month")
plt.xlabel("Visitor")

plt.show()
```

```python
fig, axs = plt.subplots(4,1, figsize=(12,8))

axs[0].bar(df8['arrival_date_month'], df8['adults'])
axs[1].plot(df8['arrival_date_month'], df8['children'], color='red', linewidth=2)
axs[2].plot(df8['arrival_date_month'], df8['babies'], color='red', linewidth=2)
axs[3].plot(df8['adults'], df8['children'], color='red', linewidth=2)

# plt.tight_layout()
plt.legend()
plt.show()
```
