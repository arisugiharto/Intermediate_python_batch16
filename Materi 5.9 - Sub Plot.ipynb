```python
from datetime import datetime
import matplotlib.dates as mpl_dates
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("data_sub.csv")
data
```

```python
data.head()
```

```python
ages = data["Age"]
all_dev = data["All_Devs"]
py = data["Python"]
js = data["JavaScript"]
```

```python
fig, ax1 = plt.subplots()
fig, ax2 = plt.subplots()
```

```python
fig, ax = plt.subplots()
ax.plot(ages, py, label = "Python")
ax.plot(ages, js, label = "JavaScript")
ax.plot(ages, all_dev, label = "All Devs")

ax.legend()

ax.set_title("Median Salary (USD) by Age")
ax.set_xlabel("Age")
ax.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows = 2, ncols = 1)

ax1.plot(ages, py, label = "Python")
ax1.plot(ages, all_dev, label = "All Devs")

ax1.legend()
ax1.set_title("Py Dev vs All Dev")
ax1.set_xlabel("Age")
ax1.set_ylabel("Median Salary (USD)")

ax2.plot(ages, js, label = "JavaScript")
ax2.plot(ages, all_dev, label = "All Devs")

ax2.legend()
ax2.set_title("Js Dev vs All Dev")
ax2.set_xlabel("Age")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```
