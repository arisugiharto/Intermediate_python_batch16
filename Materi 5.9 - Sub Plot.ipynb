```python
fig, ax = plt.subplots()
ax.plot(ages, py, label = "Python")
ax.plot(ages, js, label = "JavaScript")
ax.plot(ages, all_dev, label = "All Devs")

ax.legend()

ax.set_title("Median Salary (USD) by Age")
ax.set_xlabel("Age")
ax.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

```python
#load data
df = pd.read_csv('tesla.csv')
df['date'] = pd.to_datetime(df['date'])
df
```

# Price Insight

```python
#plot
plt.figure(figsize=(15,8))
plt.plot(df['date'], df['close'], color='red', linewidth=2)
plt.title('Tesla Stock Price',  fontweight='bold')
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Price in USD', fontweight='bold')
plt.grid()
plt.show()
```

# Volume Insight

```python
#grouping and adding up the volume by year
vol=(df.groupby(df['date'].dt.strftime('%Y'))['volume'].sum())
vol=vol.to_frame()
vol['date'] = vol.index
vol
```

```python
#plot
fig, ax = plt.subplots(figsize=(15,8))
ax.bar(vol['date'], vol['volume'])
ax.set_title('Tesla Stock Volume',  fontweight='bold')
ax.set_xlabel('Year', fontweight='bold')
ax.set_ylabel('Volume in USD e+10', fontweight='bold')
plt.show()
```
