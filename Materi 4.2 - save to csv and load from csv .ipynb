# Save to CSV

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("My Uber Drives - 2016.csv")
data
```

```python
data["DISTANCE"] = np.where(data["MILES*"] >5, 'Longtrip', 'Short Trip')
data
```

```python
data["YEAR"]=np.array("2021")
data
```

```python
data["YEAR"] = ("2020")
data
```

```python
data["TIME_CAT"] = np.where(data["START_DATE*"]>"1/06/2016 14:00:00", "New_Trip", "Old_Trip")
data
```

```python
data.to_csv("uber_baru.csv")
```

```python
#pada saat load data yg sudah di save, secara otomatis system akan menyimpan indexing (0,1,2,3,dst)/ auto create
# SYstem juga akan menampilkan kolom "UNNAMED" pada saat load data

load = pd.read_csv("uber_baru.csv")
load
```

```python
#Jika kita tidak ingin menampilkan atau menyimpan index unnamed tersebut, maka gunakan index=false
data.to_csv("uber_baru_noindex.csv", index = False)
```
