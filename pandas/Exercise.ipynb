1. In the My Uber drives - 2016 dataset, find all trips that is greater than 10 miles and originated from New York and Morrisville

```python
uber
```

```python
uber['START*'].value_counts()
```

```python
filter1 = uber.loc[uber['MILES*'] > 10]
filter1
```

```python
combine = filter1[filter1['START*'].isin(['New York', 'Morrisville'])]
combine
```

```python
summary = uber.loc[uber['MILES*'] > 10][uber.loc[uber['MILES*'] > 10]['START*'].isin(['New York', 'Morrisville'])]
summary
```

### Exercise 5

1. Using My Uber drives -2016, create a new column with the following condition:\
    a) >10    : Long Trip\
    b) 5-10   : Medium Trip\
    c) <5     : Short Trip

```python
df_new = pd.read_csv("My Uber Drives - 2016.csv")
df_new
```

```python
import numpy as np
df_new['CATEGORIES*'] = np.where(df_new['MILES*'] > 10, 'Long Trip', np.where(df_new['MILES*'] > 5, 'Medium Trip', 'Short Trip'))
df_new
```
