# 5. Replace Missing Values

__`fillna()`__ function in Pandas is used to replace missing values with another values. 

Missing values can be replaced by:
1. Special value
2. Aggregate value, such as mean, median, etc

## Replacing with scalar

```python
import numpy as np
import pandas as pd
```

```python
new_df = pd.DataFrame({"col_a": [1, 2, 4, 1, np.nan, np.nan, 5],
                      "col_b": [3,7,np.nan, 9, None, 5, 8],
                      "col_c": ["a","?","x","y","--", np.nan, "r"],
                      "col_d": [True, True,np.nan, None, False, True, False]})
new_df
```

```python
df8 = new_df.replace({"?": np.nan, "--": np.nan})
df8
```

```python
df_withoutmissing=df8
```

```python
df_withoutmissing
```

```python
df_withoutmissing.fillna(3)
```

```python
df_withoutmissing=df8
```

```python
df_withoutmissing.iloc[:, 2].mode()
```

```python
df_withoutmissing.iloc[:, 2].mode()[2]
```

```python
df_withoutmissing.iloc[:,2]=df_withoutmissing.iloc[:,2].fillna(df_withoutmissing.iloc[:,2].mode()[1])
```

```python
df_withoutmissing
```

```python
df_withoutmissing.isnull().sum()
```

Untuk merubah __`col_a`__ menjadi rata2 (mean)

```python
df_withoutmissing
```

```python
df_withoutmissing.iloc[:,0].mean()
```

```python
df_withoutmissing.iloc[:,0]=df_withoutmissing.iloc[:,0].fillna(df_withoutmissing.iloc[:,0].mean())
```

```python
df_withoutmissing
```

```python
df_withoutmissing.isnull().sum()
```

```python
df_withoutmissing
```

__`ffill`__: mengisi nilai `NaN` dengan nilai dari nilai sebelumnya

```python
df_withoutmissing.fillna(method = 'ffill')
```

__`bfill`__: mengisi nilai `NaN` dengan nilai dari nilai sesudahnya

```python
df_withoutmissing.fillna(method = 'bfill')
```
