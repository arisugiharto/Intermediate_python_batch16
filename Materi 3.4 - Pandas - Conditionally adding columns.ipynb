Di sini kita akan memakai `library numpy` dengan menggunakan `np.where(data["MILES*])` Perbedaan dengan menggunakan `if else `adalah jika `if else` hanya berpengaruh pada beberapa row saja sedangkan `np.wher`e berpengaruh pada `seluruh data sets`

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("My Uber Drives - 2016.csv")
data
```

```python
# Mile >5 --> long trip
# Mile <5 --> Short trip
# data["DISTANCE"] : akan menambahkan kolom baru bernama `DISTANCE`

data["DISTANCE"] = np.where(data["MILES*"] >5, 'Longtrip', 'Short Trip')
data
```

Cara menghitung jumlah (total), "Long Trip", "Short Trip pada table d atas"

```python
data["DISTANCE"].value_counts()
```

```python
# Untuk menampilkan data yang sama pada table

data["YEAR"]=np.array("2021")
data
```

```python
# Atau dapat juga menggunakan command di bawah ini:
data["YEAR"] = ("2020")
data
```

```python
#untuk melihat jumlah yg unik dari suatu data
data["START*"].value_counts()
```
