2. concentenate the following arrays 
**[[0, 1, 3], [5, 7, 9]]** and **[[0, 2, 4], [6, 8, 10]]**

```python
matrix2 = np.array([[0,1,3],
                   [5,7,9]])
matrix3 = np.array([[0,2,4],
                    [6,8,10]])
matrix4 = np.concatenate((matrix2, matrix3), axis = 0)
matrix4
```

```python
matrix4 = np.concatenate((matrix2, matrix3), axis = 1)
matrix4
```
