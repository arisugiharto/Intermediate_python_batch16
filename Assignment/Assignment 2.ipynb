    - Embarked 's'
    - Age 'Mean'
    - Cabin 'Mode'


```python
data
```

```python
data2=data
```

```python
data2
```

```python
#  Mengganti nilai kosong di kolom 'embarked' dengan nilai 's'
data2.iloc[:, 11].mode()
```

```python
data2.iloc[:, 11]=data2.iloc[:, 11].fillna(data2.iloc[:, 11].mode()[0])
```

```python
data2
```

```python
data2.info()
```

```python
data2.isnull().sum()
```

```python
data2
```

```python
data2.iloc[:, 5].mean()
```

```python
# Mengganti nilai kosong di kolom 'age' dengan nilai mean()
data2.iloc[:, 5]=data2.iloc[:, 5].fillna(data2.iloc[:, 5].mean())
```

```python
data2
```

```python
# Mengganti nilai kosong di kolom 'kabin' dengan nilai mode()
#reff: https://stackoverflow.com/questions/42789324/how-to-pandas-fillna-with-mode-of-column

data2['Cabin'].fillna(data2['Cabin'].mode()[0], inplace = True)
```

```python
data2
```

```python
data2.isnull().sum()
```
