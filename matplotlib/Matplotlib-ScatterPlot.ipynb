# SCATTER PLOT

![](matplotlib/scatter.png)

It is used to visualize the relationship of two values and see whether it is correlated or not

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.scatter(x, y)

plt.tight_layout()
plt.show()
```

```python
#setting size
plt.scatter(x, y, s=100)

plt.tight_layout()
plt.show()
```

```python
#change color

plt.scatter(x, y, s=100, c='#A760FF')

plt.tight_layout()
plt.show()
```

```python
#change marker

plt.scatter(x, y, s=100, c='green', marker='v')

plt.tight_layout()
plt.show()
```

```python
#add edge color

plt.scatter(x, y, s=100, c='green', edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()
plt.show()
```

Have multiple colors and sizes. enable us to have more datasets

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 4, 3, 5, 6, 7, 1, 1, 8, 5, 6, 9, 3, 5, 6, 2, 6, 2, 8, 7]
```

```python
plt.scatter(x, y, s=100, c=colors, edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()
plt.show()
```

We can use color map

```python
plt.scatter(x, y, s=100, c=colors, cmap='hot', edgecolor='black',
            linewidth=1, alpha=0.75)

plt.tight_layout()
plt.show()
```

cmaps can be seen here: https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

#Add color bar legend

```python
plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black',
            linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()
plt.show()
```

```python
#sizes

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 4, 3, 5, 6, 7, 1, 1, 8, 5, 6, 9, 3, 5, 6, 2, 6, 2, 8, 7]
sizes = [117, 234, 343, 435, 556, 417, 145, 591, 518, 225, 286, 419, 373, 425, 246, 542, 196, 255, 468, 417]
```

```python
plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', edgecolor='black',
            linewidth=1, alpha=0.75)

cbar= plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()
plt.show()
```

# Use real world data

```python
data = pd.read_csv('data_scatter.csv')
data.head()
```

```python
view_count = data['view_count'] # X
likes = data['likes'] # Y
ratio = data['ratio']
```

```python
plt.style.use('seaborn')

plt.scatter(view_count, likes, edgecolor='black',
            linewidth=1, alpha=0.75)

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

```python
plt.scatter(view_count, likes, edgecolor='black',
            linewidth=1, alpha=0.75)

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

### use the color for ratio

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=1)
cbar=plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.xlabel('Total Likes')

plt.tight_layout()
plt.show()
```
