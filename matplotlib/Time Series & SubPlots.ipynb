```python
#lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
df = pd.read_csv('tesla.csv')
df['date'] = pd.to_datetime(df['date'])
df
```

# Price Insight

```python
#plot
plt.figure(figsize=(15,8))
plt.plot(df['date'], df['close'], color='red', linewidth=2)
plt.title('Tesla Stock Price',  fontweight='bold')
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Price in USD', fontweight='bold')
plt.grid()
plt.show()
```

# Volume Insight

```python
#grouping and adding up the volume by year
vol=(df.groupby(df['date'].dt.strftime('%Y'))['volume'].sum())
vol=vol.to_frame()
vol['date'] = vol.index
vol
```

```python
#plot
fig, ax = plt.subplots(figsize=(15,8))
ax.bar(vol['date'], vol['volume'])
ax.set_title('Tesla Stock Volume',  fontweight='bold')
ax.set_xlabel('Year', fontweight='bold')
ax.set_ylabel('Volume in USD e+10', fontweight='bold')
plt.show()
```

```python
fig, axs = plt.subplots(4,1, figsize=(12,8))

axs[0].bar(vol['date'], vol['volume'])
axs[1].plot(df['date'], df['close'], color='red', linewidth=2)
axs[2].plot(df['date'], df['close'], color='red', linewidth=2)
axs[3].plot(df['date'], df['close'], color='red', linewidth=2)

# plt.tight_layout()
plt.show()
```
