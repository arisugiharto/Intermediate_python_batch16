## Histogram

```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
ages = [18,19,21,25,26,26,30,32,38,45,55]
```

If we do survey to some people, it might be able for us to plot in bar graph. But what if we have 100 or more people? That's why we need histogram.
Histogram allows us to create bins for our data

```python
plt.hist(ages)

plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.show()
```

```python
plt.hist(ages, bins = 5) #bins = range data

plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.show()
```

### Add Edge Color

```python
plt.hist(ages, bins = 5, edgecolor = "k") # edgecolor = untuk memisahkan antar bins dalam table. "k" = black

plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.show()
```

#### We can pass our own list of value (bins). It allows us to have control

```python
#Plot 10 Yrs of difference

ages = [18,19,21,25,26,26,30,32,38,45,55]
bins = [10,20,30,40,50,60] # Membuat range data ages dalam satuan bins tertentu
```

```python
plt.hist(ages, bins = bins, edgecolor = "k")

plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.show()
```

#### We can exclude a data

```python
ages = [18,19,21,25,26,26,30,32,38,45,55]
bins = [20,30,40,50,60]

plt.hist(ages, bins = bins, edgecolor = "k")

plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.show()
```

## Let's use real world data

```python
data = pd.read_csv("data_histo.csv")
data
```

```python
data.info()
```

```python
ids = data["Responder_id"]
ages = data["Age"]

bins = [10,20,30,40,50,60,80,90,100]
```

```python
plt.hist(ages, bins = bins, edgecolor = "k")

plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.show()
```

```python
plt.hist(ages, bins = bins, edgecolor = "k", log = True)

plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.show()
```

#### Add Median Line

```python
median_age = 29 # median line = garis rata-rata
color = "r"

plt.hist(ages, bins = bins, edgecolor = "k", log = True) # log = Untuk meringkas data pada ylabel agar tidak terlalu banyak
plt.axvline(median_age, color = color, label = "age_media")#axvline = membuat garis median tersebut

plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.show()
```

```python
median_age = 29 # median line = garis rata-rata
color = "r"

plt.hist(ages, bins = bins, edgecolor = "k", log = True) 
plt.axvline(median_age, color = color, label = "age_media", linewidth = 2)#Linewidth = tebal garis median

plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.show()
```
