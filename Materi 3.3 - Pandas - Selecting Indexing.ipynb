## Loc

Perbedaan antara __`Loc`__ dengan __`Iloc`__ adalah pada saat pemanggilannya di mana __`Iloc`__ menggunakan indexing (angka) sedangkan __`Loc`__ menggunakan nama kolomnya

```python
#[1:5] : adalah jumlah kolom yang akan di tampilkan
#["START*", "STOP*", "MILES*"] : nama kolom yang akan di tampilkan

data.loc[1:5, ["START*", "STOP*", "MILES*"]]
```

## 2. Filtering

```python
# menampilkan data-data yang akan di filter sesuai kebutuhan
# Akan menampilkan data dengan MILES>10 dan dengan nama kolom ["END_DATE*", "MILES*", "CATEGORY*"]

filter1 = data.loc[data["MILES*"]>10, ["END_DATE*", "MILES*", "CATEGORY*"]]
filter1
```

```python
# Menampilkan seluruh data dengan miles > 10
filter2 = data.loc[data["MILES*"]>10]
filter2
```

```python
# Menampilkan data dengan start ==newyork
filter3 = data.loc[data["START*"]=="New York"]
filter3
```

```python
#isin: nilainya sama seperti '==' tp dapat memiliki beberapa nilai di dalamnya (>1 variabel)

data[data["START*"].isin(["New York", 'Gampaha'])]
```
